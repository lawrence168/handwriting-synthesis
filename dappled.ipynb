# Handwriting Synthesis

Not too long ago I couldn't figure out how a piece of marketing junk mail managed to mass-produce what looked like an custom, authentic handwritten message where the same letters were never written twice in the same way. Now I know--by using recurrent neural networks!

This notebook uses https://github.com/sjvasquez/handwriting-synthesis to synthesize handwriting and can be customized to different styles.

```python
text = ''
text__placeholder = 'Text to generate handwriting from'

bias = 0.75
bias__range = [0.0, 1.0, 0.05]

style = 4
style__choices = list(range(1, 13))

color = 'blue'
color__choices = ['blue', 'red', 'black', 'green']

# preload model
from demo import Hand
hand = Hand()
```

```python
from IPython.display import display, SVG
import io

if not text:
    text = 'Just like real handwriting!'

buf = io.StringIO()
hand.write(
    fileobj=buf,
    lines=[text],
    biases=[bias],
    styles=[style],
    stroke_colors=[color],
    stroke_widths=[1]
)
display(SVG(data=buf.getvalue()))
```
